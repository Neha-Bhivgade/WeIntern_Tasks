```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("intern_performance_dashboard_dataset.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
weekly_tasks = df.groupby("Week")["Tasks Completed"].sum()

plt.figure()
weekly_tasks.plot(kind="bar")
plt.title("Weekly Task Completion")
plt.xlabel("Week")
plt.ylabel("Total Tasks Completed")
plt.show()
```

```python
attendance = df.groupby("Intern Name")["Attendance (%)"].mean()

plt.figure()
attendance.plot(kind="bar")
plt.title("Average Attendance of Interns")
plt.xlabel("Intern Name")
plt.ylabel("Attendance (%)")
plt.show()
```

```python
domain_tasks = df.groupby("Domain")["Tasks Completed"].sum()

plt.figure()
domain_tasks.plot(kind="pie", autopct="%1.1f%%")
plt.title("Domain-wise Task Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
plt.hist(df["Performance Score"], bins=10)
plt.title("Performance Score Distribution")
plt.xlabel("Performance Score")
plt.ylabel("Frequency")
plt.show()
```

```python
top_interns = df.groupby("Intern Name")["Performance Score"].mean().sort_values(ascending=False)
top_interns.head()
```
